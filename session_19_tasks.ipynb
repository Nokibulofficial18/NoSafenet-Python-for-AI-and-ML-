### `Q-4:` Perform `groupby` based on the `"Team"` column and then perform Z Normalization on top of the below columns of each group:
1. Passes
2. Passes Completed
3. Attempted Line Breaks
4. Completed Line Breaks

You have to make a python function named `z_normalization` which takes two arguments:

1. *group:* Every group that you have created
2. *cols_to_perform:* This parameter takes a list of columns on which you have to perform the Z-Normalization.

For this problem, you have to use th `apply()` method.

$\Large Z \ - \ Normalization = \frac{X_i - \mu}{std} $

After that find out the below values for each group:
- minimum "Passess"
- maximum "Passess"
- minimum "Yellow Cards"
- maximum "Yellow cards"
- average "Yellow Cards"
- maximum "Attempted Line Breaks"
- minimum "Attempted Line Breaks"
- standardard deviation of "Attempted Line Breaks"
- average Possession

````python
# code here
import pandas as pd

fifa_df = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vT3D_x_4DS6d51LKJ7ze1sxT5WpV5uiSVOFYHLwBiGru6vFyVv5h5-83AwFjxWYiWfCDjDAaarHAV-k/pub?gid=0&single=true&output=csv")
fifa_df.sample()
fifa_df.columns
````

````python
def z_normalization(group,cols_to_perform):
  for col in cols_to_perform:
    group[col] = (group[col] - group[col].mean())/group[col].std()
  return group
````

````python
cols_to_normalize = ["Passes", "Passes Completed", "Attempted Line Breaks", "Completed Line Breaks"]
normalize_df = fifa_df.groupby("Team").apply(z_normalization,cols_to_normalize)
normalize_df.head()
````

````python
fifa_df.groupby("Team").agg({
    "Passes":["min","max"],
    "Yellow Cards": ["min", "max", "mean"],
    "Attempted Line Breaks": ["min", "max", "std"],
    "Possession (%)": "mean"

})
````

ball by ball dataset - https://drive.google.com/file/d/1-kvv_9KCSAFWcrhS9WgTxSrURkRh6GNt/view?usp=share_link





````python
import pandas as pd

url = "https://drive.google.com/uc?id=1-kvv_9KCSAFWcrhS9WgTxSrURkRh6GNt"
ipl_df = pd.read_csv(url)
ipl_df.head()
````

###`Q-5:` Find batsman in below category-
* Highest score while chasing
* Best Strike rate while chasing and have faced 100+ balls


> Chasing mean team batting in second inning

````python
# code here
chasing_df = ipl_df[ipl_df["innings"]==2]
highest_run = chasing_df.groupby(["batter","ID"])["batsman_run"].sum().max()
highest_run
````

###`Q-6` Most Successful bowler against any batsman. Find that pair of bowler and batsman.
> Most Successful in terms of dissmissal. A bowler who have dissmissed any batsman most no of times. If any two pairs have same no of dissmisal, consider runs conceded by bowler to that batsman. Those who have concede lesser runs is more successful.

````python
import pandas as pd

# Step 1: Filter only dismissals (where a wicket was taken)
dismissals_df = ipl_df[ipl_df["isWicketDelivery"] == 1]

# Step 2: Count dismissals per (bowler, batter) pair
dismissal_counts = dismissals_df.groupby(["bowler", "batter"]).size().reset_index(name="Dismissals")

# Step 3: Calculate total runs conceded per (bowler, batter) pair
# Use 'batsman_run' for batter-specific runs (ignores extras)
runs_conceded = ipl_df.groupby(["bowler", "batter"])["batsman_run"].sum().reset_index(name="Runs_Conceded")

# Step 4: Merge both
pair_stats = pd.merge(dismissal_counts, runs_conceded, on=["bowler", "batter"])

# Step 5: Sort by highest dismissals, then lowest runs conceded
sorted_pairs = pair_stats.sort_values(by=["Dismissals", "Runs_Conceded"], ascending=[False, True])

# Step 6: Get the top pair
most_successful = sorted_pairs.iloc[0]

# Step 7: Output
print("Most Successful Bowler vs Batsman Pair:")
print(f"{most_successful['bowler']} dismissed {most_successful['batter']} "
      f"{most_successful['Dismissals']} times, conceding {most_successful['Runs_Conceded']} runs.")
````

###`Q-7`: Most successful batting pair in IPL. Batting pair who have scored most runs playing together.


````python
# code here



ipl_df["batting_pair"] = ipl_df.apply(
    lambda row: tuple(sorted([row["batter"], row["non-striker"]])), axis=1
)

pair_runs = ipl_df.groupby("batting_pair")["total_run"].sum().reset_index(name="Total_Runs")


pair_runs_sorted = pair_runs.sort_values(by="Total_Runs", ascending=False)


top_pair = pair_runs_sorted.iloc[0]

print("Most Successful Batting Pair in IPL:")
print(f"{top_pair['batting_pair'][0]} & {top_pair['batting_pair'][1]} - {top_pair['Total_Runs']} runs together")
````

###`Q-8:` Make a dataframe for all batting pairs played together.
```
Batsman1 Batsman2 Runs Avg StrikeRate
```

> Just to ease this question you can count wide-balls for strike rate.

````python
# code here
import pandas as pd

# Step 1: Create consistent batting pair (alphabetically sorted to avoid duplication)
ipl_df["batting_pair"] = ipl_df.apply(
    lambda row: tuple(sorted([row["batter"], row["non-striker"]])), axis=1
)

# Step 2: Group by batting pair and sum total runs
batting_pairs_df = ipl_df.groupby("batting_pair")["total_run"].sum().reset_index(name="Total_Runs_Together")

# Step 3: Split tuple into two separate columns
batting_pairs_df[["Batter_1", "Batter_2"]] = pd.DataFrame(batting_pairs_df["batting_pair"].tolist(), index=batting_pairs_df.index)

# Step 4: Rearrange columns
batting_pairs_df = batting_pairs_df[["Batter_1", "Batter_2", "Total_Runs_Together"]]

# Optional: Sort by total runs
batting_pairs_df = batting_pairs_df.sort_values(by="Total_Runs_Together", ascending=False).reset_index(drop=True)

# Show the resulting DataFrame
print(batting_pairs_df.head())
````
