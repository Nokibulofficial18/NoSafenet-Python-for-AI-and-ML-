###`Q-8:` Team1 vs Team2 Dashbord
Create a function which will take two string(name of two teams) as input. Show win Loss record between them and player getting most player of the match award in matches between these two teams.
```
team1_vs_team2('Kolkata Knight Riders','Chennai Super Kings')
```

````python
import matplotlib.pyplot as plt

def team1_vs_team2(team1, team2):
    # Filter matches between the two teams
    matches = ipl[(((ipl['Team1'] == team1) & (ipl['Team2'] == team2)) |
                ((ipl['Team1'] == team2) & (ipl['Team2'] == team1)))]

    if matches.empty:
        print(f"No matches found between {team1} and {team2}")
        return

    # Calculate win/loss record (FIXED: shape is a property, not a function)
    team1_wins = matches[matches["WinningTeam"] == team1].shape[0]
    team2_wins = matches[matches["WinningTeam"] == team2].shape[0]
    ties = matches[matches["WinningTeam"].isna()].shape[0]  # Added tie calculation

    # Get top players (FIXED: Using value_counts correctly)
    top_player = matches['Player_of_Match'].value_counts().head(5)

    # Create dashboard
    plt.figure(figsize=(14, 6))

    # Win/Loss Pie Chart (FIXED: Using correct variable names)
    plt.subplot(1, 2, 1)
    labels = [f'{team1} Wins', f'{team2} Wins', 'Ties/No Result']
    sizes = [team1_wins, team2_wins, ties]  # Now includes ties
    colors = ['#ff9999', '#66b3ff', '#cccccc']
    plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
    plt.title(f'Head-to-Head Record\n{team1} vs {team2}', pad=20)

    # Top Player Bar Chart
    plt.subplot(1, 2, 2)
    bars = plt.bar(top_player.index, top_player.values, color='#2ca02c')
    plt.title('Top Player of Match Performers', pad=20)
    plt.xlabel('Player')
    plt.ylabel('Number of Awards')
    plt.xticks(rotation=45, ha='right')

    # Add value labels
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, height,
                f'{height}', ha='center', va='bottom')

    plt.tight_layout()

    # Print summary statistics
    print(f"\n{team1} vs {team2} Summary:")
    print(f"Total Matches Played: {len(matches)}")
    print(f"{team1} Wins: {team1_wins}")
    print(f"{team2} Wins: {team2_wins}")
    print(f"Ties/No Result: {ties}")
    print(f"\nMost 'Player of Match' Awards:")
    print(top_player.to_string(header=False))

    plt.show()

# Example usage
team1_vs_team2('Kolkata Knight Riders', 'Chennai Super Kings')
````

### `Q-9:` Find out the top 7 cities where the matches of Kolkata Knight Riders are played frequently and plot the result as bar chart.

*`.plot(kind = "bar")` can help you to plot the bar chart. Also you can learn more about this method from [here](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.html)*



````python
# code here
kkr_matches = ipl[(ipl['Team1'] == 'Kolkata Knight Riders') | (ipl['Team2'] == 'Kolkata Knight Riders')]
top_cities = kkr_matches['City'].value_counts().head(7)
plt.figure(figsize=(12, 6))
top_cities.plot(kind='bar', color='purple', alpha=0.7)

# Customizing the plot
plt.title('Top 7 Cities Where Kolkata Knight Riders Have Played Most Matches', pad=20)
plt.xlabel('City', labelpad=10)
plt.ylabel('Number of Matches', labelpad=10)
plt.xticks(rotation=45, ha='right')

# plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
````
