###`Q-6:` Make a dataframe of each team in IPL with details like - Team Name, Matches Played, Win%, Home Win%, Away Win%.
Show sorted dataframe on Win%

Replace old team name as new name before performing any tasks.
```
Delhi Daredevils ->Delhi Capitals
Kings XI Punjab -> Punjab Kings
Rising Pune Supergiants -> Rising Pune Supergiant
```

Note: Team1 represents Home team. Exclude not result matches.


````python
replace_dict = {
    'Delhi Daredevils': 'Delhi Capitals',
    'Kings XI Punjab': 'Punjab Kings',
    'Rising Pune Supergiants': 'Rising Pune Supergiant'
}

ipl_matches.replace({'Team1': replace_dict, 'Team2': replace_dict, 'WinningTeam': replace_dict}, inplace=True)
ipl_matches.tail(5)
````

````python
# code here
teams = pd.unique(ipl_matches[['Team1','Team2']].values.ravel())
data = []

for team in teams:
  total_match = ipl_matches[(ipl_matches['Team1']==team) | (ipl_matches['Team2']==team)].shape[0]
  wins = ipl_matches[ipl_matches['WinningTeam']==team].shape[0]
  home_match = ipl_matches[ipl_matches['Team1']==team].shape[0]
  home_match_win = ipl_matches[(ipl_matches['WinningTeam']==team) & (ipl_matches['Team1']==team)].shape[0]
  away_match = ipl_matches[ipl_matches['Team2']==team].shape[0]
  away_match_win = ipl_matches[(ipl_matches['WinningTeam']==team) & (ipl_matches['Team2']==team)].shape[0]

  win_pct = round(100 * wins/total_match,2)
  home_win_pct = round(100 * home_match_win / home_match,2)
  away_win_pct = round(100* away_match_win/away_match,2)
  data.append([team,total_match,win_pct,home_win_pct,away_win_pct])


team_df = pd.DataFrame(data,columns=['Team Name','Match Played', 'Win%','Home Win%','Away Win%'])

team_df = team_df.sort_values(by="Win%",ascending=False).reset_index(drop=True)

print(team_df)
````

###`Q-9:` IPL Point Table

Make a function `point_table` which take `season` as parameter and show points table in non-ascendng order of points and in ascending order of team name.

For winning - 2 Ponits;
For loosing - 0 Point
For not result both team gets 1 points.

Make dataframe which will have
`TeamName` `MatchesPlayed` `MatchesWon` `NoResult` `Points`
make `TeamName` as index.

```
season parametr should be one of these->
['2022', '2021', '2020/21', '2019', '2018', '2017', '2016', '2015', '2014', '2013', '2012', '2011', '2009/10', '2009', '2007/08']
```


Output of two Top 2 in season 2022
```
TeamName    MatchesPlayed	MatchesWon	NoResult	Points

Gujarat Titans	    16	   12	       0	     24
Rajasthan Royals	  17	   10	       0	     20

```

````python
def point_table(season):
    # Convert season to string to match data type
    season = str(season)

    # Step 1: Filter matches for the season
    season_df = ipl_matches[ipl_matches['Season'] == season]

    if season_df.empty:
        print(f"No matches found for season {season}")
        return pd.DataFrame()  # Return empty DataFrame

    # Step 2: Initialize dictionary to collect data
    teams = pd.unique(season_df[['Team1', 'Team2']].values.ravel())
    data = {team: {'MatchesPlayed': 0, 'MatchesWon': 0, 'NoResult': 0} for team in teams}

    # Step 3: Iterate over each match
    for _, row in season_df.iterrows():
        team1 = row['Team1']
        team2 = row['Team2']
        winner = row['WinningTeam']

        data[team1]['MatchesPlayed'] += 1
        data[team2]['MatchesPlayed'] += 1

        if pd.isna(winner):  # No Result
            data[team1]['NoResult'] += 1
            data[team2]['NoResult'] += 1
        else:
            data[winner]['MatchesWon'] += 1

    # Step 4: Build DataFrame
    table = []
    for team, stats in data.items():
        matches = stats['MatchesPlayed']
        wins = stats['MatchesWon']
        no_result = stats['NoResult']
        points = wins * 2 + no_result
        table.append([team, matches, wins, no_result, points])

    df_points = pd.DataFrame(table, columns=['TeamName', 'MatchesPlayed', 'MatchesWon', 'NoResult', 'Points'])

    if df_points.empty:
        return df_points

    df_points.set_index('TeamName', inplace=True)

    # Step 5: Sort: First by Points DESC, then by TeamName ASC
    df_points = df_points.sort_values(by=['Points', 'TeamName'], ascending=[False, True]).reset_index().set_index('TeamName')


    return df_points
````

````python
print(point_table(2022))
````

###`Q-10:` IPL Point Table cont.
Extend the above IPL Point Table with an extra column as `SeasonPosition`

Team below top 4 after sorting on `Points` and then on `TeamName` Will have same `SeasonPosition` as there rank. use rank function.

Teams in Top four will have `SeasonPosition` as:
```
    'Winner' - Team won final
    'Runner' - Team lost Final
    3 - Losing Team in Qualifier2
    4 - Losing Team in Eliminator
```

For changing value of pariticular cell use `df.at[row_index, col_label] = value`

Output of two Top 2 in season 2022. Your result should have all teams.
```
TeamName    MatchesPlayed	MatchesWon	NoResult	Points   SeasonPosition

Gujarat Titans	    16	   12	       0	     24         Winner
Rajasthan Royals	  17	   10	       0	     20         Runner

```

Note: If you try to chnage value of view of any dataframe a warnig will be shown. To avoid it, make a copy of the dataframe you want to change in by `df.copy()`

````python
# code here
def point_table(season):
    season = str(season)
    season_df = ipl_matches[ipl_matches['Season'] == season].copy()

    if season_df.empty:
        print(f"No matches found for season {season}")
        return pd.DataFrame()

    # Init data
    teams = pd.unique(season_df[['Team1', 'Team2']].values.ravel())
    data = {team: {'MatchesPlayed': 0, 'MatchesWon': 0, 'NoResult': 0} for team in teams}

    for _, row in season_df.iterrows():
        team1, team2 = row['Team1'], row['Team2']
        winner = row['WinningTeam']
        data[team1]['MatchesPlayed'] += 1
        data[team2]['MatchesPlayed'] += 1

        if pd.isna(winner):
            data[team1]['NoResult'] += 1
            data[team2]['NoResult'] += 1
        else:
            data[winner]['MatchesWon'] += 1

    # Build points DataFrame
    table = []
    for team, stats in data.items():
        points = stats['MatchesWon'] * 2 + stats['NoResult']
        table.append([team, stats['MatchesPlayed'], stats['MatchesWon'], stats['NoResult'], points])

    df_points = pd.DataFrame(table, columns=['TeamName', 'MatchesPlayed', 'MatchesWon', 'NoResult', 'Points'])

    # Sort: Points DESC, then TeamName ASC
    df_points = df_points.sort_values(by=['Points', 'TeamName'], ascending=[False, True]).reset_index(drop=True)
    df_points = df_points.copy()  # avoid SettingWithCopyWarning

    # Add placeholder for SeasonPosition
    df_points['SeasonPosition'] = df_points['Points'].rank(method='dense', ascending=False).astype(int)

    # --- Handle top 4 custom labels ---
    final = season_df[season_df['MatchNumber'].str.contains("Final", na=False)]
    qualifier2 = season_df[season_df['MatchNumber'].str.contains("Qualifier 2", na=False)]
    eliminator = season_df[season_df['MatchNumber'].str.contains("Eliminator", na=False)]

    if not final.empty:
        winner = final.iloc[0]['WinningTeam']
        runner = final.iloc[0]['Team1'] if final.iloc[0]['Team2'] == winner else final.iloc[0]['Team2']
        df_points.at[df_points[df_points['TeamName'] == winner].index[0], 'SeasonPosition'] = 'Winner'
        df_points.at[df_points[df_points['TeamName'] == runner].index[0], 'SeasonPosition'] = 'Runner'

    if not qualifier2.empty:
        loser_q2 = qualifier2.iloc[0]['Team1'] if qualifier2.iloc[0]['WinningTeam'] == qualifier2.iloc[0]['Team2'] else qualifier2.iloc[0]['Team2']
        df_points.at[df_points[df_points['TeamName'] == loser_q2].index[0], 'SeasonPosition'] = 3

    if not eliminator.empty:
        loser_elim = eliminator.iloc[0]['Team1'] if eliminator.iloc[0]['WinningTeam'] == eliminator.iloc[0]['Team2'] else eliminator.iloc[0]['Team2']
        df_points.at[df_points[df_points['TeamName'] == loser_elim].index[0], 'SeasonPosition'] = 4

    return df_points.set_index("TeamName")
````
